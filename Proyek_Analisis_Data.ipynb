### Membaca Dataset dan menggabungkannya

```python
# Misalkan setiap variabel dataset adalah DataFrame yang sudah dibaca dari file CSV
Aotizhongxin = pd.read_csv("Dataset\\PRSA_Data_Aotizhongxin_20130301-20170228.csv")
Changping = pd.read_csv("Dataset\\PRSA_Data_Changping_20130301-20170228.csv")
Dingling = pd.read_csv("Dataset\\PRSA_Data_Dingling_20130301-20170228.csv")
Dongsi = pd.read_csv("Dataset\\PRSA_Data_Dongsi_20130301-20170228.csv")
Guanyuan = pd.read_csv("Dataset\\PRSA_Data_Guanyuan_20130301-20170228.csv")
Gucheng = pd.read_csv("Dataset\\PRSA_Data_Gucheng_20130301-20170228.csv")
Huairou = pd.read_csv("Dataset\\PRSA_Data_Huairou_20130301-20170228.csv")
Nongzhanguan = pd.read_csv("Dataset\\PRSA_Data_Nongzhanguan_20130301-20170228.csv")
Shunyi = pd.read_csv("Dataset\\PRSA_Data_Shunyi_20130301-20170228.csv")
Tiantan = pd.read_csv("Dataset\\PRSA_Data_Tiantan_20130301-20170228.csv")
Wanliu = pd.read_csv("Dataset\\PRSA_Data_Wanliu_20130301-20170228.csv")
Wanshouxigong = pd.read_csv("Dataset\\PRSA_Data_Wanshouxigong_20130301-20170228.csv")

# Membaca setiap DataFrame menjadi list
dfs = [Aotizhongxin, Changping, Dingling, Dongsi, Guanyuan,
       Gucheng, Huairou, Nongzhanguan, Shunyi, Tiantan,
       Wanliu, Wanshouxigong]

# Menggabungkan (merge) DataFrame secara vertikal menggunakan concat dengan axis=0
merged_data = pd.concat(dfs, axis=0)

merged_data
```

```python
# Buang kolom 'wd', 'station', dan 'timestamp'
columns_to_drop = ['wd', 'station', 'timestamp']
merged_data_numeric = merged_data.drop(columns=columns_to_drop, axis=1)

# Hitung matriks korelasi untuk variabel numerik
numeric_corr_matrix = merged_data_numeric.corr()

# Heatmap untuk matriks korelasi numerik
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriks Korelasi Variabel Numerik')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pilih kolom yang akan diikutsertakan dalam analisis
selected_columns = ['TEMP', 'PRES', 'DEWP', 'RAIN', 'PM2.5']

# Filter hanya kolom yang diperlukan
environmental_data = merged_data[selected_columns]

# Visualisasi matriks korelasi
correlation_matrix = environmental_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriks Korelasi antara Faktor Lingkungan dan Konsentrasi PM2.5')
plt.show()

# Visualisasi scatter plot untuk analisis individu
plt.figure(figsize=(15, 10))
for i, column in enumerate(selected_columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.scatterplot(x=column, y='PM2.5', data=environmental_data)
    plt.title(f'Scatter Plot {column} vs PM2.5')

plt.tight_layout()
plt.show()

# Analisis pengaruh arah dan kecepatan angin terhadap dispersi polutan udara
plt.figure(figsize=(15, 8))
plt.subplot(1, 2, 1)
sns.scatterplot(x='wd', y='PM2.5', data=merged_data)
plt.title('Scatter Plot Arah Angin vs PM2.5')

plt.subplot(1, 2, 2)
sns.scatterplot(x='WSPM', y='PM2.5', data=merged_data)
plt.title('Scatter Plot Kecepatan Angin vs PM2.5')

plt.tight_layout()
plt.show()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Apa penyumbang Konsentrasi Gas terbesar di Seluruh stasiun, dan Stasiun Mana yang merupakan penyumbang Konsentrasi gas terbesar?

```python
# Hitung rata-rata konsentrasi gas di seluruh stasiun
average_gas_concentration = merged_data[['SO2', 'NO2', 'CO', 'O3']].mean()

# Visualisasi rata-rata konsentrasi gas di seluruh stasiun menggunakan diagram batang
plt.figure(figsize=(10, 6))
average_gas_concentration.plot(kind='bar', color=['red', 'green', 'blue', 'orange'])
plt.title('Rata-Rata Konsentrasi Gas di Seluruh Stasiun')
plt.xlabel('Gas')
plt.ylabel('Rata-Rata Konsentrasi')
plt.show()
```

```python
# Hitung total konsentrasi gas untuk setiap stasiun
total_gas_concentration_by_station = merged_data.groupby('station')[['SO2', 'NO2', 'CO', 'O3']].sum()

# Hitung total konsentrasi gas di seluruh stasiun
total_gas_concentration_all_stations = total_gas_concentration_by_station.sum(axis=1)

# Temukan stasiun dengan total konsentrasi gas terbesar
station_with_highest_total_pollution = total_gas_concentration_all_stations.idxmax()

# Visualisasi total konsentrasi gas di seluruh stasiun menggunakan diagram batang
plt.figure(figsize=(12, 6))
total_gas_concentration_all_stations.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Total Konsentrasi Gas di Seluruh Stasiun')
plt.xlabel('Stasiun')
plt.ylabel('Total Konsentrasi Gas')
plt.show()

print(f"Stasiun dengan total konsentrasi gas terbesar: {station_with_highest_total_pollution}")
```

### Pertanyaan 2: Pada waktu berapa Konsentrasi gas CO terbesar pada stasiun Wanshouxigong

```python
import pandas as pd
import matplotlib.pyplot as plt


# Ekstrak jam dari timestamp
merged_data['hour'] = merged_data['timestamp'].dt.hour

# Pilih kolom yang akan divisualisasikan (contoh: PM2.5)
selected_column = 'CO'
selected_station = 'Wanshouxigong'

# Filter data untuk stasiun 'Dongsi'
dongsi_data = merged_data[merged_data['station'] == selected_station]

# Temukan jam dengan konsentrasi maksimum untuk stasiun 'Dongsi'
max_hour_per_station = dongsi_data.groupby(['hour'])[selected_column].idxmax()
max_hour_data = dongsi_data.loc[max_hour_per_station]

# Visualisasi jam dengan konsentrasi maksimum
plt.figure(figsize=(15, 8))
plt.plot(max_hour_data['hour'], max_hour_data[selected_column], label=selected_station)

plt.title(f'Jam dengan Konsentrasi Maksimum {selected_column} di Stasiun {selected_station}')
plt.xlabel('Jam')
plt.ylabel(f'Konsentrasi {selected_column}')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Konversi kolom timestamp ke tipe datetime
merged_data['timestamp'] = pd.to_datetime(merged_data['timestamp'])

# Opsi 2: Ekstrak tahun
merged_data['year'] = merged_data['timestamp'].dt.year
selected_column = 'CO'
selected_value = 'year'

# Filter data berdasarkan stasiun
selected_station = 'Wanshouxigong'
selected_data = merged_data[merged_data['station'] == selected_station]

# Temukan jam atau tahun dengan konsentrasi maksimum untuk stasiun tertentu
max_value_per_station = selected_data.groupby([selected_value])[selected_column].idxmax()
max_value_data = selected_data.loc[max_value_per_station]

# Visualisasi jam atau tahun dengan konsentrasi maksimum
plt.figure(figsize=(15, 8))
plt.plot(max_value_data[selected_value], max_value_data[selected_column], label=selected_station)

plt.title(f'Tahun dengan Konsentrasi Maksimum {selected_column} di Stasiun {selected_station}')
plt.xlabel(selected_value.capitalize())
plt.ylabel(f'Konsentrasi {selected_column}')
plt.legend()
plt.show()
```
